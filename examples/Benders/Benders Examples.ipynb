# Example 1
## Structure

![alt text](https://github.com/welchalisandra/BendersDecomposition/raw/master/Type1.png)

## Master Problem
* Objective:     $ Min\ 2y $
* Constraint:    $ y\geq0 $

## Subproblem
* Objective: $Min\ 2x_1+3x_2$
* Constraints: 

   $2x_1-x_2+3y\geq4$

   $x_1+2x_2+y\geq3$
   
   $y\geq0$
   
   $x_1\geq0,\   x_2\geq0$

# Example 2
## Structure

![alt text](https://github.com/welchalisandra/BendersDecomposition/raw/master/Type3.png)

## Master Problem
* Objective:     $ Min\ 150x_1+230x_2+260x_3 $
* Constraints:    

   $ x_1+x_2+x_3 \leq 500 $
   
   $x_1\geq0,\ x_2\geq0, \ x_3\geq0$

## Subproblem 1
* Objective: $Min\ -\frac{1}{3}(170w_1-238y_1+150w_2-210y_2+36w_3+10w_4)$
* Constraints: 

   $3x_1+y_1-w_1\geq200$

   $3.6x_2+y_2-w_2\geq 240$
   
   $w_3\leq6000$
   
   $w_1\geq0, \ w_2\geq0, \ w_3\geq0, \ w_4\geq0$
   
   $y_1\geq0, \ y_2\geq0$
   
   $x_1\geq0, \ x_2\geq0, \ x_3\geq0$
   
## Subproblem 2
* Objective: $Min\ -\frac{1}{3}(170w_5-238y_3+150w_6-210y_4+36w_7+10w_8)$
* Constraints: 

   $2.5x_1+y_3-w_5\geq200$

   $3x_2+y_4-w_6\geq240$
   
   $w_7+w_8\leq20x_3$
   
   $w_7\leq6000$
   
   $w_5\geq0, \ w_6\geq0, \ w_7\geq0, \ w_8\geq0$
   
   $y_3\geq0, \ y_4\geq0$
   
   $x_1\geq0, \ x_2\geq0, \ x_3\geq0$
   
   
## Subproblem 3
* Objective: $Min\ -\frac{1}{3}(170w_9-238y_5+150w_10-210y_6+36w_11+10w_12)$
* Constraints: 

   $2x_1+y_5-w_9\geq200$

   $2.4x_2+y_6-w_10\geq240$
   
   $w_11+w_12\leq16x_3$
   
   $w_11\leq6000$
   
   $w_9\geq0, \ w_10\geq0, \ w_11\geq0, \ w_12\geq0$
   
   $y_5\geq0, \ y_6\geq0$
   
   $x_1\geq0, \ x_2\geq0, \ x_3\geq0$
